```
import pandas as pd

df = pd.read_csv("/datasets/employee-compensation/Employee_Compensation.csv")
```

```
filter_col = [col for col in df.columns if 'Code' not in col]
```

```
df_filtered = df[filter_col]
```

```
df_filtered = df_filtered[df_filtered['Salaries'] > 0]
```

### Exploratory Data Analysis

```
salary_yr = df_filtered.groupby(['Year']).mean()['Salaries'].reset_index()
```

```
salary_yr
```

```
top_10_dep = df_filtered.groupby(['Department','Year']).mean().sort_values(['Year','Department','Salaries']).nlargest(10, 'Department')['Salaries'].reset_index()
```

```
top_10_dep
```

```
import plotly.express as px
fig = px.scatter(top_10_dep, x="Year", y="Salaries",facet_row='Department')
fig.show()
```
